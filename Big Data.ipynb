# Q15: Creating a new dataframe for when close price is above 300 and volume is above average

```python
vol = tesla["volume"] # make vol the volume 
filter2 = (vol >= vol.mean()) & (close >= 300) #create a second filter where it onyl takes days with above avergae volume and a close price of above $300
print("The number of days when close price is above $300 and the average number of days is above average:", filter2.sum(), "days") #print statement
tesla_2 = tesla[filter2]
```
